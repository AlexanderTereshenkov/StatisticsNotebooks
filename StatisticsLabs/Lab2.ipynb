```python
def get_freq(nums, a, b, k):
    freq = [0] * (k + 1)
    step = (b - a) / k
    x = []
    first_elem = a
    x.append(first_elem)
    while first_elem < b:
        first_elem += step
        x.append(first_elem)
    for num in nums:
        for i in range(len(x) - 1):
            if x[i] <= num < x[i+1]:
                freq[i] += 1
                break
    freq = [x / len(nums) for x in freq]
    chi_squared = 0
    for p in freq:
        if p != 0:
            chi_squared += ((p - 0.1) ** 2) / 0.1
    #chi_squared *= len(nums)
    return [freq, x, 1 - stats.chi2.cdf(chi_squared, k-1)]
```
