```python
from scipy.stats import f
```

```python
data = [[3, 7, 3, 6, 6, 7, 6, 3, 8, 3],
        [6, 5, 7, 4, 9, 4, 3, 2, 7, 8],
        [8, 6, 3, 2, 7, 8, 6, 9, 3, 8],
        [4, 7, 7, 8, 6, 4, 5, 8, 4, 7],
        [6, 2, 6, 6, 8, 9, 7, 6, 8, 1]]
a_mean = False
b_mean = False
#m = len(data)
#k = len(data[i])

#Среднее по всей таблице
temp_sum = 0
for i in range(len(data)):
    for j in range(len(data[i])):
        temp_sum += data[i][j]
avg_x_all_data = temp_sum / (len(data) * len(data[0]))

#Среднее по столбцам
avg_column = []
#Среднее по строкам
avg_row = []
k = len(data[0])
m = len(data)

for i in range(k):
    t_sum = 0
    for j in range(m):
        t_sum += data[j][i]
    avg_column.append(t_sum / m)

for i in range(m):
    t_sum = 0
    for j in range(k):
        t_sum += data[i][j]
    avg_row.append(t_sum / k)

print(avg_column)
print(avg_row)

#Временная суммы
temp_sum_1 = 0
for i in range(k):
    for j in range(m):
        temp_sum_1 += (data[j][i] - avg_column[i]) ** 2
#print(temp_sum_1)
temp_sum_2 = 0
for x_j in avg_row:
    temp_sum_2 += (x_j - avg_x_all_data) ** 2

#Дисперсия фактора случайности
s_0 = (1 / ((k - 1) * (m - 1))) * (temp_sum_1 - k * temp_sum_2)
print(s_0, "Дисперсия фактора случайности")

#Выборочная дисперсия, включающая фактор A
temp_sum = 0
for x_i in avg_column:
    temp_sum += (x_i - avg_x_all_data) ** 2
s_na = (m / (k - 1)) * temp_sum
print(s_na, "Выборочная дисперсия фактора А")

p = 0.95
f_critical = f.ppf(p, k - 1, (k - 1) * (m - 1))
s_critical = s_na / s_0
print("Фактор А:")
print(s_critical, " ", f_critical)
if(s_critical > f_critical):
    a_mean = True
    print("Фактор А значимый")
else:
    print("Фактор А не значимый")

#Выборочная дисперсия, включающая фактор B
temp_sum = 0
for x_j in avg_row:
    temp_sum += (x_j - avg_x_all_data) ** 2
s_nb = (k / (m - 1)) * temp_sum
print(s_nb, "ВЫборочная дисперсия фактора B")

p = 0.95
f_critical = f.ppf(p, m - 1, (k - 1) * (m - 1))
s_critical = s_nb / s_0
print("Фактор B:")
print(s_critical, " ", f_critical)
if(s_critical > f_critical):
    b_mean = True
    print("Фактор B значимый")
else:
    print("Фактор B не значимый")

s_a = 0
s_b = 0
S = 0

# Оценка дисперсий фактора A или B при их значимости
if a_mean:
    s_a = (s_na - s_0) / (k)
    print(s_a, "Дисперсия фактора А")
elif b_mean:
    s_b = (s_nb - s_0) / (m)
    print(s_b,  "Дисперсия фактора В")

#Оценка генеральной дисперсии
if not(a_mean) and b_mean:
    S = ((k-1) * s_na + (k-1)*(m-1)*s_0) / ((k-1) + (k-1)*(m-1))
elif not(b_mean) and a_mean:
    S = ((m-1) * s_nb + (k-1)*(m-1)*s_0) / ((m-1) + (k-1)*(m-1))
elif not(a_mean) and not(b_mean):
    S = ((k-1) * s_na + (m-1) * s_nb + (k-1)*(m-1)*s_0) / ((k-1) + (m-1) + (k-1)*(m-1))
print(S, "Оценка генеральной дисперсии")
```
