```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
numbers = np.random.normal(loc=0, scale=1, size=1000)
```

```python
plt.hist(numbers, bins=100)
plt.show()
```

```python
#random 20% with random.uniform
nums_copy_1 = numbers.copy()
error = [random.uniform(-0.2, 0.2) for x in range(len(nums_copy_1))]
nums_copy_1 = [nums_copy_1[i] * (1 + error[i]) for i in range(len(nums_copy_1))]
plt.hist(nums_copy_1, bins=100)
plt.show()
```

```python
nums_copy_2 = numbers.copy()
for i in range(len(nums_copy_2)):
    left = nums_copy_2[i] - (nums_copy_2[i] * 0.2)
    rigth = nums_copy_2[i] + (nums_copy_2[i] * 0.2)
    nums_copy_2[i] = random.uniform(left, rigth)
plt.hist(nums_copy_2, bins=100)
plt.show()
```

```python
n = []
for i in range(10000):
    n.append(random.uniform(-5, 5))
plt.hist(n, bins=100)
plt.show()
```
