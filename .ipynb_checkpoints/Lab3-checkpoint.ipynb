```python
def get_freq(nums, a, b, k):
    freq = [0] * (k + 1)
    step = (b - a) / k
    x = []
    first_elem = a
    while first_elem <= b:
        x.append(first_elem)
        first_elem += step
    if(len(x) < len(freq) and x[-1] < b):
        x.append(b)
    for num in nums:
        for i in range(len(x) - 1):
            if x[i] <= num < x[i+1]:
                freq[i] += 1
                break
    freq = [x / len(nums) for x in freq]
    print(len(freq), "AAAAAAA")
    return [freq, x]
```

```python
def T_normal(n):
    k1 = 1
    k2 = 2
    k3 = 5
    k4 = 4
    numbers = []
    for i in range(n):
        T = random.uniform(0, k1) + random.uniform(0, k2) + random.uniform(0, k3) + random.uniform(0, k4)
        numbers.append(T)
    return numbers
```

```python
k1 = 1
k2 = 2
k3 = 5
k4 = 4

K = k1 + k2 + k3 + k4

M = k1 / 2 + k2 / 2 + k3 / 2 + k4 / 2
D = (k1 ** 2) / 12 + (k2 ** 2) / 12 + (k3 ** 2) / 12 + (k4 ** 2) / 12
S = D ** 0.5

N = [10, 20, 50, 100, 200, 500, 1000]
print(M, D, S)
```

```python
for n in N:
    
    nums = T_normal(n)
    arr = get_freq(nums, 0, K, 12)
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_subplot()
    
    x = arr[1]
    y = arr[0]
    
    m = sum(nums) / len(nums)
    
    x_t = np.linspace(0, K, 100)
    y_t = (1 / ((2 * math.pi) ** 0.5 * S)) * math.e ** ((-1 / 2) * ((x_t - M) / (S)) ** 2)
    
    #print(m)
    
    ax.bar(x, y, align='edge', width=0.8)
    ax.grid()
    
    plt.plot(x_t, y_t)
    plt.show()

```
